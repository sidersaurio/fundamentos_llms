## Cuantizado

```python
example_weights = [
    [0.12, -0.47, 0.33, 0.08],
    [-0.25, 0.19, -0.41, 0.56],
    [0.72, -0.15, 0.05, -0.32]
]
```

```python
n_bits = 8
max_int = 2**(n_bits - 1) - 1

def clamp(x, low, high):
    return max(low, min(high, x))
```

```python
alpha = max(abs(w) for row in example_weights for w in row)
```

```python
scale = alpha / max_int
```

```python
quantized = []
for row in example_weights:
    q_row = []
    for x in row:
        x_q = clamp(round(x / scale), -max_int, max_int)
        q_row.append(int(x_q))
    quantized.append(q_row)
```

```python
quantized
```

## Decuantizado

```python
dequantized = [[x_q * scale for x_q in row] for row in quantized]
dequantized
```

```python
dequantized = [[round(x_q * scale, 2) for x_q in row] for row in quantized]
dequantized
```
